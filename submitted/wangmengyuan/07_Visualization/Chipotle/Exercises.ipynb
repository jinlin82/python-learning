### Step 5. Create a histogram of the top 5 items bought

```python
x=chipo.item_name
y=Counter(x)
z=pd.DataFrame.from_dict(y,orient='index')#è½¬æ¢ä¸ºæ•°æ®æ¡†
z=z[0].sort_values(ascending=False).head(5)#ä»Žå¤§åˆ°å°æŽ’åºï¼Œæ‰¾åˆ°5ä¸ªæœ€å¤šitem
z.plot(kind='bar')
plt.xlabel('item')#è®¾ç½®æ ‡ç­¾å’Œé¢˜ç›®
plt.ylabel('item_price')
plt.title('histogram of the top 5 items bought')
plt.show
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
chipo.item_price=[float(value[1:-1]) for value in chipo.item_price]#åˆ©ç”¨æ•°ç»„åˆ‡ç‰‡åŽ»é™¤$ç¬¦å·ï¼Œå°†objectåž‹è½¬ä¸ºfloat
chipo.dtypes
orders=chipo.groupby('item_price').sum()
orders
plt.scatter(x=orders.index ,y=orders['quantity'])
```
