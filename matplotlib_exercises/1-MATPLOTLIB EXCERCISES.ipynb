```python
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
```

## Excercise 1

1. Create a figure object called fig using plt.figure()
2. Use add_axes to add an axis to the figure canvas at [0,0,1,1]
3. Plot (x,y) on that axes and set the labels and titles

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Title')
```

## Excercise 2

## Excercise 3

Create the plot by adding two axes to a figure obeject at [0,0,1,1] and ]0.2,0.5,.4,.4]

```python
fig = plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2 =fig.add_axes([0.2,0.5,.4,.4])
```

## Now use x,y and z arrays to recreate the plot below. Notice the xlimits and ylimits on the inserted plot:

```python
ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title("Zoom")
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

fig
```

## Excercise 4

Use plt.subplots(nrows=1, ncols=2) to create the plot below.

```python
plt.subplots(nrows=1,ncols=2)
```

## Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'b',lw=4)
axes[1].plot(x,z,'r--',lw=3)
```

## Resize the plot by adding the figsize() argument in plt.subplots() are copying and passing the previous code

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(10,2))
axes[0].plot(x,y,'b',lw=4)
axes[1].plot(x,z,'r--',lw=3)
```
